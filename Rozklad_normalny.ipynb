```python
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
```

Dane

```python
# 1. zbiór
mean_1 = 3
std_dev_1 = 1
num_of_samples_1 = 100
data_1 = np.random.normal(mean_1, std_dev_1, num_of_samples_1)
# 2. zbiór
mean_2 = 3
std_dev_2 = 1
num_of_samples_2 = 10000
data_2 = np.random.normal(mean_2, std_dev_2, num_of_samples_2)
```

Mediana, moda, zakres, kwartyle

```python
# 1. zbiór
median_1=np.median(data_1)
q0_1 = np.quantile(data_1, 0.0)
q1_1 = np.quantile(data_1, 0.25)
q3_1 = np.quantile(data_1, 0.75)
q4_1 = np.quantile(data_1, 1.0)
range_1=q4_1-q0_1
print("Zbiór 1.:")
print(f"Średnia: {mean_1}")
print(f"Mediana (Q2): {median_1}")
print(f"Min: {q0_1}")
print(f"Q1: {q1_1}")
print(f"Q3: {q3_1}")
print(f"Max: {q4_1}")
print(f"Zakres: {range_1}")
# 2. Zbiór
median_2=np.median(data_2)
q0_2 = np.quantile(data_2, 0.0)
q1_2 = np.quantile(data_2, 0.25)
q3_2 = np.quantile(data_2, 0.75)
q4_2 = np.quantile(data_2, 1.0)
range_2=q4_2-q0_2
print("Zbiór 2.:")
print(f"Średnia: {mean_2}")
print(f"Mediana (Q2): {median_2}")
print(f"Min: {q0_2}")
print(f"Q1: {q1_2}")
print(f"Q3: {q3_2}")
print(f"Max: {q4_2}")
print(f"Zakres: {range_2}")
```

Wykresy z wyliczonymi danymi statystycznymi

```python
# 1. zbiór
plt.hist(data_1, bins=10)
plt.axvline(x = mean_1, color='red', label = 'Średnia')
plt.axvline(x = median_1, color='green', label = 'Mediana (Q2)')
plt.axvline(x = q0_1, color='yellow', label = 'Min')
plt.axvline(x = q1_1, color='black', label = 'Q1')
plt.axvline(x = q3_1, color='black', label = 'Q3')
plt.axvline(x = q4_1, color='yellow', label = 'Max')
plt.legend(loc='upper right')
plt.show()
```

```python
# 2. zbiór
plt.hist(data_2, bins=20)
plt.axvline(x = mean_2, color='red', label = 'Średnia')
plt.axvline(x = median_2, color='green', label = 'Mediana (Q2)')
plt.axvline(x = q0_2, color='yellow', label = 'Min')
plt.axvline(x = q1_2, color='black', label = 'Q1')
plt.axvline(x = q3_2, color='black', label = 'Q3')
plt.axvline(x = q4_2, color='yellow', label = 'Max')
plt.legend(loc='upper right')
plt.show()
```

Wnioski:
- im większa liczność próby, tym bardziej zbliżone są do siebie miediana i średnia, które w rozkładzie normalnym są sobie tożsame.
- wykres / histogram jest bardziej "smukły" przy większej próbie i najbardziej przypomina wykres dla rozkładu normalnego (krzywą Gaussa)
